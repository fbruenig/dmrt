```python
import diffTools
import numpy as np
import matplotlib.pyplot as plt
import pylange
```

```python
# Do doublewell simulation and analyse
radii=np.linspace(-1,1,num=100)
difft = diffTools.DiffTools()
inp = pylange.lange(P=6,N=0.1,m=0.01,G=1, dt=0.01)
dists,ptpxOD,cts,ptpxOD0,ptpxOD1 = difft.compute(inp[:,:2], radii=radii, mode='ptpxbins',verb=True)

inp = pylange.lange(P=6,N=0.1,m=10,G=1, dt=0.01)
dists,ptpx,cts,ptpx0,ptpx1 = difft.compute(inp[:,:2], radii=radii, mode='ptpxbins',verb=True)
plt.plot(dists,ptpxOD,label="overdamped")
plt.plot(dists,ptpx,label="inertial")
plt.legend()
plt.show()

plt.plot(dists,ptpxOD0,label="overdamped, L-R")
plt.plot(dists,ptpx0,label="inertial, L-R")
plt.plot(dists,ptpxOD1,label="overdamped, R-L")
plt.plot(dists,ptpx1,label="inertial, R-L")
plt.legend()
plt.show()
```
